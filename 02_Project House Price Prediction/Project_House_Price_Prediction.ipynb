## 1️⃣ Dataset Loading

```python
import pandas as pd

# Load dataset
df = pd.read_csv("HousePricePrediction.csv")

# Display first 10 rows
df.head(10)
```

## 2️⃣ Data Exploration

```python
# Shape of dataset (rows, columns)
print("Shape:", df.shape)
```

```python
# Data types of all columns
print("\nData Types:\n", df.dtypes)
```

```python
# Summary statistics of numerical features
df.describe()
```

## 3️⃣ Data Cleaning

```python
# Check missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Handle missing values (example: fill numeric with median, categorical with mode)
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

print("Shape after cleaning:", df.shape)
```

## 4️⃣ Feature Selection

```python
print(df.columns)
```

```python
# Drop unnecessary identifier column
df = df.drop("Id", axis=1)
```

```python
# Preview selected features
df.head()
```

## 5️⃣ Data Preprocessing

```python
# Convert categorical variables into numerical form using one-hot encoding
df_encoded = pd.get_dummies(df, drop_first=True)
```
